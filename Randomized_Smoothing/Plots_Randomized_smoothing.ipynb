```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# Plots des train et test losses

## 25% corrupted images

```python
txt = '/content/drive/MyDrive/Dauphine/M2/Projets DS/Adversarial_attacks/smoothing-master/model_output_dir/cifar10/resnet50/noise_0.25/log.txt'
df_noise_25 = pd.read_csv(txt, '\t')
df_noise_25.head()
```

```python
train_acc = df_noise_25['train acc'].tolist()[-1]
test_acc = df_noise_25['test acc'].tolist()[-1]

print("The train accuracy for a model trained on data with a 25% corruption is {}".format(train_acc))
print("The test accuracy for a model trained on data with a 25% corruption is {}".format(test_acc))
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(data = df_noise_25, x='epoch', y='train loss', label='train loss')
sns.lineplot(data = df_noise_25, x='epoch', y='testloss', label='test loss')

plt.xlabel('Number of epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss of a randomized smoothed classifier trained on images of noise 0.25')
plt.show()
```

# Accuracy for a smoothed classifier of model trained with different proportions of corrupted images for test set of 50% corrputed images

Certify over 10 000 images, skipped 100 --> 1000 analyzed images 10%

## Model trained on 25% corrupted images

```python
txt = '/content/drive/MyDrive/Dauphine/M2/Projets DS/Adversarial_attacks/smoothing-master/data/certify/cifar10/resnet50/noise_0.25/test/sigma_0.50.txt'
df_noise_25_sigma_50 = pd.read_csv(txt,'\t')
df_noise_25_sigma_50.head()
```

```python
acc = df_noise_25_sigma_50['correct'].sum()/len(df_noise_25_sigma_50['correct'])* 100
print("The model certifies 10% of the size of the test set which is 10 000.")
print("The accuracy of a model trained on 25% corrupted data for 50% corrupted data is {}%".format(acc))
```

The accuracy is not good because the test data was too corrupted for the model. Indeed, it was trained on 25% corrupted data.
